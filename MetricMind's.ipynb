1.   data cleaning on'followersCount' column --> 'cleaned_column1'

```python
import pandas as pd

# the values in the followersCount are strings , convert them into numeric form

df['followersCount'] = df['followersCount'].str.replace('"', '').str.strip()
df['cleaned_column1'] = pd.to_numeric(df['followersCount'], errors='coerce')

# drop the Nan values
df = df.dropna(subset=['cleaned_column1'])

# Extract the first digit of each number
first_digits = df['cleaned_column1'].astype(int).abs().astype(str).str[0].astype(int)

frequency = first_digits.value_counts().sort_index()

# Calculate the percentage for each digit
percentage = (frequency / frequency.sum()) * 100

print(frequency)
```

```python
#the percentage distribution of (1-9)
percentage
```

```python
import matplotlib.pyplot as plt


plt.bar(frequency.index, frequency.values, color='skyblue', edgecolor='black')
plt.title("First Digit Frequency Distribution")
plt.xlabel("First Digit")
plt.ylabel("Frequency")
plt.xticks(range(1, 10))

plt.show()
```

As we can see , the percentage distribution for the 'followersCount' column , for the first digit of each value matches the benfords law.
Even the graph also shows the same plot .

::->

Hence we can Validate that the 'followersCount' column follows Bendford's Law.

2. Data Cleaning on ['tweetId']  --> 'cleaned_column2'

```python
import pandas as pd
import numpy as np

digits = range(1, 10)
# Clean and convert tweetId to numeric
df['tweetId'] = df['tweetId'].str.replace('"', '').str.strip()
df['cleaned_column2'] = pd.to_numeric(df['tweetId'], errors='coerce')

# Drop NaN values
df = df.dropna(subset=['cleaned_column2'])

# Get first digits
tweetid_digits = [int(str(int(id)).replace(",", "")[0]) for id in df['cleaned_column2']]

# Calculate observed and expected counts
observed_counts_tweetid = [tweetid_digits.count(digit) for digit in range(1, 10)]
total_records_tweetid = len(tweetid_digits)
expected_counts_tweetid = [int(total_records_tweetid * np.log10(1 + 1 / digit)) for digit in range(1, 10)]
```

```python
import seaborn as sns

sns.lineplot(x=digits, y=observed_counts_tweetid, label="Observed")
sns.lineplot(x=digits, y=expected_counts_tweetid, label="Expected")

plt.xlabel("Leading Digit")
plt.ylabel("Frequency")
plt.title("Digit Distribution(tweet_id)")
plt.legend()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df['cleaned_column1'] = pd.to_numeric(df['cleaned_column1'], errors='coerce')
df['cleaned_column2'] = pd.to_numeric(df['cleaned_column2'], errors='coerce')

df = df.dropna(subset=['cleaned_column1', 'cleaned_column2'])

# We define a function to extract the first digit
def leading_digit(series):
    return series.astype(str).str.lstrip('-0.').str[0].astype(int)

df['digit_col1'] = leading_digit(df['cleaned_column1'])
df['digit_col2'] = leading_digit(df['cleaned_column2'])


freq_df = pd.DataFrame({
    'cleaned_column1': df['digit_col1'].value_counts(normalize=True).sort_index(),
    'cleaned_column2': df['digit_col2'].value_counts(normalize=True).sort_index()
}).reindex(range(1, 10), fill_value=0)


plt.figure(figsize=(8, 5))
sns.heatmap(freq_df.T, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Frequency'}, fmt=".2f")
plt.title("Heatmap of Leading Digit Frequencies (Benford's Law)")
plt.xlabel("Leading Digit")
plt.ylabel("Column")
plt.show()

```

The about heatmap  shows how closely the distribution of leading digits in your dataset (cleaned_column1 and cleaned_column2) aligns with Benfords Law.

Both  [cleaned_column1]  and [cleaned_column2]  exhibit a strong concentration of leading digits toward 1 and 2, closely aligning with Benfords Law.
This suggests that the numerical data in these columns follows expected natural patterns.
